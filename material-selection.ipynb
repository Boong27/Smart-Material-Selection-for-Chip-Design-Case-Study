```python
# ✅ 1. Cài đặt thư viện nếu cần
!pip install pandas matplotlib

# ✅ 2. Upload file từ máy bạn (Google Colab)
from google.colab import files
uploaded = files.upload()  # Chọn file: project3_material_selection.csv

# ✅ 3. Đọc dữ liệu
import pandas as pd

file_name = "project3_material_selection.csv"
df = pd.read_csv(file_name)

print("📊 Data Loaded:")
display(df)

# ✅ 4. Tách dữ liệu theo từng use case
fast_charger_criteria = [
    "Switching Frequency (kHz)",
    "Gate Charge (nC)",
    "On-Resistance (mΩ)",
    "Size Index (1=Large, 0=Small)"
]

ev_inverter_criteria = [
    "Voltage Handling (V)",
    "Thermal Conductivity (W/m·K)",
    "Switching Loss (0–1)",
    "Max Operating Temp (°C)"
]

logic_ic_criteria = [
    "CMOS Compatibility (1/0)",
    "Cost (0–1)",
    "Yield (%)",
    "Toolchain Support (1/0)"
]

# ✅ 5. Hàm vẽ biểu đồ cho từng use case
import matplotlib.pyplot as plt

def plot_use_case(title, criteria_list):
    sub_df = df[df["Criteria"].isin(criteria_list)]
    for i, row in sub_df.iterrows():
        plt.figure(figsize=(6, 4))
        plt.bar(["Si", "GaN", "SiC"], row[1:], color=["gray", "blue", "green"])
        plt.title(f"{title}: {row['Criteria']}")
        plt.ylabel("Value")
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.tight_layout()
        plt.show()

# ✅ 6. Vẽ cho từng tình huống ứng dụng
print("🔌 Fast Charger (GaN-focused use case):")
plot_use_case("Fast Charger", fast_charger_criteria)

print("🚗 EV Power Inverter (SiC-focused use case):")
plot_use_case("EV Inverter", ev_inverter_criteria)

print("⚙️ Logic IC / Consumer Chip (Si-focused use case):")
plot_use_case("Logic IC", logic_ic_criteria)
```
